1. ĞŸĞ¾Ğ´Ğ±ĞµÑ€Ğ¸Ñ‚Ğµ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚ÑŒ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ (eta) Ğ¸ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹

```python
import pandas as pd
import numpy as np
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
```

```python
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
# def calc_mae(y, y_pred):
#     err = np.mean(np.abs(y - y_pred))
#     return err

def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
n = X.shape[0]

eta = 2e-2      # Ğ’Ğ½Ñ‘Ñ Ğ¸Ğ·Ğ¼ĞµĞ½ĞµĞ½Ğ¸Ñ Ğ² ÑˆĞ°Ğ³, ÑƒĞ²ĞµĞ»Ğ¸Ñ‡Ğ¸Ğ» Ñ 0.01 Ğ´Ğ¾ 0.02
n_iter = 2500   # Ğ§Ğ¸ÑĞ»Ğ¾ Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹ ÑƒĞ²ĞµĞ»Ğ¸Ñ‡Ğ¸Ğ» Ğ´Ğ¾ 2500, Ğ´Ğ°Ğ»ÑŒÑˆĞµ Ğ¸Ğ·Ğ¼ĞµĞ½ĞµĞ½Ğ¸Ñ Ğ¿Ñ€Ğ¾Ğ¸ÑÑ…Ğ¾Ğ´ÑÑ‚ Ğ² 8 Ğ·Ğ½Ğ°ĞºĞµ Ğ¿Ğ¾ÑĞ»Ğµ Ğ·Ğ°Ğ¿ÑÑ‚Ğ¾Ğ¹

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

2*. Ğ’ ÑÑ‚Ğ¾Ğ¼ ĞºĞ¾Ğ´Ğµ Ğ¼Ñ‹ Ğ¸Ğ·Ğ±Ğ°Ğ²Ğ»ÑĞµĞ¼ÑÑ Ğ¾Ñ‚ Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹ Ğ¿Ğ¾ Ğ²ĞµÑĞ°Ğ¼, Ğ½Ğ¾ Ñ‚ÑƒÑ‚ ĞµÑÑ‚ÑŒ Ğ¾ÑˆĞ¸Ğ±ĞºĞ°, Ğ¸ÑĞ¿Ñ€Ğ°Ğ²ÑŒÑ‚Ğµ ĞµĞµ

```python
n = X.shape[0]

eta = 1e-2 
n_iter = 200

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
#     for k in range(W.shape[0]):
#         W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    # Ğ˜Ğ—ĞœĞ•ĞĞ•ĞĞ˜Ğ¯
    W -= eta * (1/n * 2 * np.dot(X.T, y_pred - y))      # ĞÑˆĞ¸Ğ±ĞºĞ° Ğ±Ñ‹Ğ»Ğ° Ğ² Ñ‚Ğ¾Ğ¼, Ñ‡Ñ‚Ğ¾ Ğ¿Ñ€Ğ¸ Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ‡Ğ½Ğ¾Ğ¼ Ğ¿ĞµÑ€ĞµĞ¼Ğ½Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¸ Ğ½Ğµ ÑƒÑ‡Ğ¸Ñ‚Ñ‹Ğ²Ğ°Ğ»ÑÑ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€ Ğ´Ğ²ÑƒÑ… Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ†
                                                            # ĞÑƒĞ¶Ğ½Ğ¾ Ğ±Ñ€Ğ°Ñ‚ÑŒ Ğ½Ğµ Ğ²ÑÑ Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ†Ñƒ Ğ¥, Ğ° Ñ‚Ğ¾Ğ»ÑŒĞºĞ¾ Ğ²Ñ‚Ğ¾Ñ€Ğ¾Ğ¹ ÑÑ‚Ğ¾Ğ»Ğ±ĞµÑ†                                                       
    # Ğ˜Ğ—ĞœĞ•ĞĞ•ĞĞ˜Ğ¯
    #
    if i % 10 == 0:
      print(f'Iteration â„–{i}: W_new = {W}, MSE = {round(err,2)}')
```

3*. Ğ’Ğ¼ĞµÑÑ‚Ğ¾ Ñ‚Ğ¾Ğ³Ğ¾, Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ Ğ·Ğ°Ğ´Ğ°Ğ²Ğ°Ñ‚ÑŒ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹, Ğ·Ğ°Ğ´Ğ°Ğ¹Ñ‚Ğµ Ğ´Ñ€ÑƒĞ³Ğ¾Ğµ ÑƒÑĞ»Ğ¾Ğ²Ğ¸Ğµ Ğ¾ÑÑ‚Ğ°Ğ½Ğ¾Ğ²Ğ° Ğ°Ğ»Ğ³Ğ¾Ñ€Ğ¸Ñ‚Ğ¼Ğ° - ĞºĞ¾Ğ³Ğ´Ğ° Ğ²ĞµÑĞ° Ğ¿ĞµÑ€ĞµÑÑ‚Ğ°ÑÑ‚ Ğ¸Ğ·Ğ¼ĞµĞ½ÑÑ‚ÑŒÑÑ Ğ¼ĞµĞ½ÑŒÑˆĞµ Ğ¾Ğ¿Ñ€ĞµĞ´ĞµĞ»ĞµĞ½Ğ½Ğ¾Ğ³Ğ¾ Ğ¿Ğ¾Ñ€Ğ¾Ğ³Ğ°  ğœ– .

```python
n = X.shape[0]

eta = 1e-2

# Ğ·Ğ°Ğ´Ğ°Ğ´Ğ¸Ğ¼ Ğ½Ğ°Ñ‡Ğ°Ğ»ÑŒĞ½ÑƒÑ Ñ€Ğ°Ğ·Ğ½Ğ¸Ñ†Ñƒ Ğ²ĞµÑĞ¾Ğ² Ğ±Ğ¾Ğ»ÑŒÑˆĞ¸Ğ¼ Ñ‡Ğ¸ÑĞ»Ğ¾Ğ¼
weight_dist = np.inf

# ĞºÑ€Ğ¸Ñ‚ĞµÑ€Ğ¸Ğ¹ ÑÑ…Ğ¾Ğ´Ğ¸Ğ¼Ğ¾ÑÑ‚Ğ¸ (Ñ€Ğ°Ğ·Ğ½Ğ¸Ñ†Ğ° Ğ²ĞµÑĞ¾Ğ², Ğ¿Ñ€Ğ¸ ĞºĞ¾Ñ‚Ğ¾Ñ€Ğ¾Ğ¹ Ğ°Ğ»Ğ³Ğ¾Ñ€Ğ¸Ñ‚Ğ¼ Ğ¾ÑÑ‚Ğ°Ğ½Ğ°Ğ²Ğ»Ğ¸Ğ²Ğ°ĞµÑ‚ÑÑ)
min_weight_dist = 1e-6

# Ğ¼Ğ°ĞºÑĞ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹
max_iter = 1e4

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

iter_num = 0  # ÑÑ‡ĞµÑ‚Ñ‡Ğ¸Ğº Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹
while weight_dist > min_weight_dist and iter_num < max_iter:
    y_pred = np.dot(X, W)
    dQ = 2 / n * np.dot(X.T, y_pred - y)
    new_W = W - eta * dQ
    weight_dist = np.linalg.norm(new_W - W, ord=2)
    err = calc_mse(y, y_pred)
    W = new_W.copy()
    
    if iter_num % 100 == 0:
      print(f'Iteration #{iter_num}:  weights={W}, MSE={round(err, 2)}')
    
    iter_num += 1
```
